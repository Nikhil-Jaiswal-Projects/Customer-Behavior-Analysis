```python
data["Review Rating"] = data.groupby("Category")["Review Rating"].transform(lambda x: x.fillna(x.median()))
```

```python
data.isnull().sum()
```

```python
data.columns = data.columns.str.lower()
data.columns = data.columns.str.replace(' ','_')
data = data.rename(columns={"purchase_amount_(usd)":"purchase_amount"})
# convert column name upper to full lower case to easyly use it
# replace space with _ ('underscore')
# correct the column name
```

```python
data.columns
```

```python
# Create a column age_group 

labels = ["Young Adult","Adult","Middle-aged","Senior"]
data["age_group"] = pd.qcut(data["age"],q=4,labels = labels)

# qcut is make q = 4 it mean it make four age group by distribution 
# it split age into four diffrent age_group ny distribution and label it
```

```python
data[['age','age_group']].head(10)
```

```python
# create column of purchase_frequency_days
```

```python
data["frequency_of_purchases"].unique()
```

```python
frequency_maping = {
   'Fortnightly':14, 
   'Weekly':7, 
   'Annually':365, 
   'Quarterly':90, 
   'Bi-Weekly':14,
   'Monthly':30, 
   'Every 3 Months':90 
}

data["purchase_frequency_days"] = data["frequency_of_purchases"].map(frequency_maping)

# map each frequency and replace it with number using map and create new column name ===> "purchase_frequency_days"
```

```python
data[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
# both columns are same so drop on ecolumn = "promo_code_used"

data = data.drop('promo_code_used',axis=1)
```
